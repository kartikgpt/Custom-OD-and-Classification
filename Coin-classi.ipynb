```python
import numpy as np
import cv2
import os
import random
import matplotlib.pyplot as plt
import pickle
```

```python
DIRECTORY = '/content/drive/MyDrive/DataSet/train'
CATEGORIES = ['Five', 'One', 'Ten', 'Two']
```

```python
data = []

for category in CATEGORIES:
  folder = os.path.join(DIRECTORY, category)
  label = CATEGORIES.index(category)
  for img in os.listdir(folder):
    img_path = os.path.join(folder,img)
    img_array = cv2.imread(img_path)
    data.append([img_array, label])
```

```python
len(data)
```

```python
random.shuffle(data)
```

```python
data[0]
```

```python
X = []
y = []

for features, labels in data:
  X.append(features)
  y.append(labels)
```

```python
X = np.array(X)
y = np.array(y)
```

```python

len(y)
```

```python
pickle.dump(X,open('X.pkl','wb'))
pickle.dump(y,open('y.pkl','wb'))
```

```python
X = X/255
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
y = to_categorical(y,10)
```

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
```

```python
model = Sequential()

model.add(Conv2D(filters = 64, kernel_size = (3,3),input_shape = (200,200,3), activation = 'relu'))
model.add(MaxPooling2D((2,2)))



model.add(Conv2D(128, kernel_size = (3,3), activation = 'relu'))
model.add(MaxPooling2D((2,2)))

model.add(Dropout(0.5))
model.add(Flatten())

model.add(Dense(256,input_shape = (200,200,3), activation = 'relu'))
model.add(Dense(10, activation = 'softmax'))
```

```python
import tensorflow as tf
opt = tf.keras.optimizers.RMSprop(
    learning_rate=0.001, rho=0.9, momentum=0.0, epsilon=1e-07, centered=False,
    name='RMSprop'
)
model.compile(optimizer = opt, loss = 'categorical_crossentropy', metrics = ['accuracy'])
```

```python
model.fit(X,y,epochs = 100, validation_split = 0.1, steps_per_epoch=10, batch_size = 32)
```
